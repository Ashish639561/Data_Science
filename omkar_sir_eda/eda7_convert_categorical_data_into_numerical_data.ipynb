$method-2:np.where$

- np.where() is applicable for binary condition
- which mean it is applicable only for two lables
- np.where(condition,Truevalue,Falsevalue)
- for example case_status has two lables 
- condition: <col_values> == 'Certified'
- True value: Replace all certified values with 0
- False value: Replace all denied value with 1

```python
file_path = r'C:\Users\LENOVO\Documents\Data_Science\Naresh_it\Data Science\Batch october\datasets\omkar_sir_datasets\Visadataset.csv'
visa_df = pd.read_csv(file_path)
visa_df
```

```python
cond = visa_df['case_status'] == 'Certified'
np.where(cond,0,1)
visa_df['case_status'] = np.where(cond,0,1)
```

```python
visa_df
```

$LabelEncoder$

- LableEncoder is a method from sklearn
- under sklearn we have sub modules
- One of the submodule: Preprocessing
- any sklearn packages we have only 3 steps 
- step-1: Read the package
- step-2: Save the package
- step-3: apply the transform

```python
file_path = r'C:\Users\LENOVO\Documents\Data_Science\Naresh_it\Data Science\Batch october\datasets\omkar_sir_datasets\Visadataset.csv'
visa_df = pd.read_csv(file_path)
visa_df
```

```python
#step-1
from sklearn.preprocessing import LabelEncoder

#step-2
le = LabelEncoder()

#step-3
visa_df['case_status'] = le.fit_transform(visa_df['case_status'])
visa_df
```
